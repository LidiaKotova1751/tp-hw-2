#### 1.1. **(1,5 балла)** Реализация класса `Ingredient`

Класс `Ingredient` представляет собой базовую единицу рецепта – отдельный продукт или компонент блюда (мука, яйца, масло и т.д.). Он нужен для того, чтобы хранить информацию о каждом продукте, его количестве и единицах измерения.

Этот класс должен иметь следующую структуру:
1. Метод `__init__`, инициализирующий три атрибута:
    - `name` – название, строка
    - `quantity` – количество, float
    - `unit` – единица измерения, строка (г, кг, мл, шт)
2. Атрибут `quantity` должен быть реализован через `@property`:
    - геттер возвращает значение
    - сеттер проверяет, что значение положительное; если нет – выбрасывает `ValueError` с сообщением `"Количество должно быть положительным"`
    - сеттер приводит значение к `float`
3. Метод `__str__` – возвращает строку вида: "Мука: 500.0 г".
4. Метод `__repr__` – возвращает строку вида: `"Ingredient('Мука', 500.0, 'г')"`
5. Метод `__eq__` – два ингредиента считаются равными, если у них совпадают `name` и `unit` (количество не важно). Поскольку в рецепте не должно быть двух записей одного ингредиента, такие записи в дальнейшем нужно будет объединять

````python
# Ваш код здесь
class Ingredient:
    def __init__(self, name: str, quantity: float, unit: str):
        self.name = name
        self.quantity = quantity
        self.unit = unit

    @property
    def quantity(self):
        return self._quantity
    
    @quantity.setter
    def quantity(self, value):
        float_val = float(value)
        if float_val <= 0:
            raise ValueError('Количество должно быть положительным')
        self._quantity = float_val

    def __str__(self):
        return f'{self.name}: {self.quantity} {self.unit}'
    
    def __repr__(self):
        return f'Ingredient({self.name!r}, {self.quantity}, {self.unit!r})'
    
    def __eq__(self, other):
        if isinstance(other, Ingredient):
            return self.name == other.name and self.unit == other.unit
        return False
````

#### 1.2. **(2 балла)** Реализация класса `Recipe`

Класс `Recipe` представляет собой рецепт одного блюда – например, пиццы Маргариты. Он нужен для того, чтобы объединить название блюда с набором ингредиентов, которые для него требуются, а также предоставить методы для работы с этими ингредиентами

Этот класс должен иметь следующую структуру:
1. Метод `__init__`, инициализирующий два атрибута:
    - `title` – название блюда, строка
    - `ingredients` – список объектов `Ingredient`
2. Метод `add_ingredient(ingredient: Ingredient)` – добавляет ингредиент в рецепт. Если ингредиент с таким названием и единицей измерения уже есть – увеличивает его количество, а не создает дубликат
3. Статический метод `is_valid_ratio(ratio)` (через `@staticmethod`) – возвращает `True`, если `ratio` является числом и больше нуля, иначе `False`
4. Метод `scale(ratio: float)` – возвращает новый объект `Recipe`, в котором количество каждого ингредиента умножено на `ratio`. Исходный рецепт не изменяется
5. Метод `__len__` – возвращает количество уникальных ингредиентов в рецепте
6. Метод `__str__` – возвращает название блюда и список ингредиентов в читаемом виде


````python
# Ваш код здесь
class Recipe:
    def __init__(self, title: str, ingredients: list[Ingredient] | None = None):
        self.title = title
        self.ingredients = []

        if ingredients is not None:
            for item in ingredients:
                self.add_ingredient(item)
    
    def add_ingredient(self, ingredient: Ingredient):
        for exis in self.ingredients:
            if exis == ingredient:
                exis.quantity += ingredient.quantity
                return

        self.ingredients.append(Ingredient(ingredient.name, ingredient.quantity, ingredient.unit))
    
    @staticmethod
    def is_valid_ratio(ratio):
        return isinstance(ratio, (int, float)) and ratio > 0
    
    def scale(self, ratio: float):
        scaled_ingredients = [Ingredient(ing.name, ing.quantity * ratio, ing.unit) for ing in self.ingredients]

        return Recipe(self.title, scaled_ingredients)
    
    def __len__(self):
        return len(self.ingredients)
    
    def __str__(self):
        lines = [f'Рецепт: {self.title}']
        for ing in self.ingredients:
            lines.append(f' - {ing}')
        return '\n'.join(lines)
````

#### 1.3. **(1,5 балла)** Реализация класса `ShoppingList`

Реалиуйте класс `ShoppingList`. Он представляет собой список покупок, который включает ингредиенты из одного или нескольких рецептов, и нужен для того, чтобы автоматически суммировать все необходимые продукты перед походом в магазин, учитывая, что разные рецепты могут использовать одни и те же ингредиенты (например, мука нужна и для пиццы Маргарита, и для пиццы 4 Сыра, и для любой классической пиццы)

Этот класс должен иметь следующую структуру:
1. Метод `__init__`, инициализирующий один атрибут:
    - `_items` – внутренний список для хранения ингредиентов с привязкой к рецепту (можно хранить кортежами):
        - `ingredient` – объект класса `Ingredient`
        - `recipe_title` – строка, название рецепта, из которого добавлен этот ингредиент
2. Метод `add_recipe(recipe: Recipe, portions: float)` – добавляет рецепт в список покупок:
    - Если `portions` <= 0 – выбрасывает исключение `ValueError` с сообщением "Количество порций должно быть положительным"
    - Масштабирует рецепт на указанное количество порций (`recipe.scale(portions)`)
    - Для каждого ингредиента из масштабированного рецепта добавляет в `_items` кортеж `(ingredient, recipe.title)`
3. Метод `remove_recipe(title: str)` – удаляет все ингредиенты, относящиеся к рецепту с указанным названием:
    - Удаляет из `_items` все элементы, у которых второй элемент кортежа (`recipe_title`) равен `title`
    - Если рецепта с таким названием нет – ничего не происходит
4. Метод `get_list()` – возвращает итоговый список ингредиентов для покупки. Рекомендуется организовать в виде словаря:
    - Пройти по всем кортежам в `_items`
    - Извлечь `ingredient` из кортежа и создать ключ `(ingredient.name, ingredient.unit)`
    - Если ключ уже есть в словаре – прибавить `ingredient.quantity` к существующему значению
    - Если ключа нет – создать запись с `ingredient.quantity`
    - После обработки всех элементов преобразовать словарь в список объектов `Ingredient` – для каждой пары `(name, unit)`, `quantity` из словаря создать `Ingredient(name, quantity, unit)`
    - Отсортировать полученный список по названию ингредиента и вернуть его
5. Метод `__add__(self, other: ShoppingList)` – объединяет два списка покупок в один и возвращает его. Исходные списки при этом не изменяются


````python
# Ваш код здесь
class ShoppingList:
    def __init__(self):
        self._items = []
    
    def add_recipe(self, recipe, portions: float):
        if portions <= 0:
            raise ValueError("Количество порций должно быть положительным")
        
        scaled_recipe = recipe.scale(portions)

        for ingredient in scaled_recipe.ingredients:
            self._items.append((ingredient, recipe.title))

    def remove_recipe(self, title: str):
        self._items = [item for item in self._items if item[1] != title]

    def get_list(self):
        summary = {}

        for ingredient, _ in self._items:
            key = (ingredient.name, ingredient.unit)

            if key in summary:
                summary[key] += ingredient.quantity
            else:
                summary[key] = ingredient.quantity
        
        res = []

        for (name, unit), quantity in summary.items():
            res.append(Ingredient(name, quantity, unit))

        res.sort(key=lambda ing: ing.name)

        return res
    
    def __add__(self, other: ShoppingList):
        if not isinstance(other, ShoppingList):
            return NotImplemented
        
        new_shoppinglist = ShoppingList()
        
        new_shoppinglist._items = self._items + other._items

        return new_shoppinglist
````

#### 1.4. **(1 балл)** Реализация класса `DietaryRecipe`

Класс `DietaryRecipe` является наследником `Recipe` и представляет рецепт с указанием диетической категории (например, `"веган"`, `"без глютена"`, `"кето"`).

Этот класс должен иметь следующую структуру:

1. Метод `__init__` – принимает `title`, `diet_type` (строка, диетическая категория) и опциональный `ingredients`. Вызывает `__init__` родительского класса через `super()`
2. Переопределённый метод `scale(ratio: float)` – возвращает новый объект **`DietaryRecipe`** (а не `Recipe`) с тем же `diet_type`. Использует `super().scale()` для получения масштабированных ингредиентов
3. Переопределённый метод `__str__` – добавляет к строке родителя префикс с диетической категорией, например: `"[веган] Пицца Маргарита"`


````python
# Ваш код здесь
class DietaryRecipe(Recipe):
    def __init__(self, title, diet_type, ingredients = None):
        super().__init__(title, ingredients)
        self.diet_type = diet_type
    
    def scale(self, ratio: float):
        scaled_recipe = super().scale(ratio)

        return DietaryRecipe(scaled_recipe.title, self.diet_type, scaled_recipe.ingredients)
    
    def __str__(self):
        base_str = super().__str__()

        return base_str.replace('Рецепт:', f'Рецепт: [{self.diet_type}]', 1)
````
